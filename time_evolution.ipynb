```python
import numpy as np
import finch_functions
from math import pi
```

```python
finch_e = finch_functions.finchEvolver(N_finches=2)
```

```python
N_gen = 4

theta_i = np.array([pi/2,pi/2,pi/2,pi/2,pi/2,pi/2,pi/2,pi/2])
theta_intra = [[pi/10,pi/10],[pi/10,pi/10]]
theta_inter = [pi/4, pi/4]

preferred_characteristics = np.array([1,1,1,1,1,1,1,1])
multiplier = 0*np.array([1,1,1,1,1,1,1,1])

shots = 50
```

```python
fdata = finch_e.multi_generation_evolution_just_data(N_gen, theta_i, theta_intra, theta_inter, preferred_characteristics, multiplier, shots)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generate random data for three time series
time = np.arange(0, 5, 1)
for j in range(4):
    plt.plot(time,[i[j] for i in fdata],linewidth=3,marker='o',label='F1')
for j in range(4,8):
    plt.plot(time,[i[j] for i in fdata],linewidth=3,marker='o',label='F2')

# add labels and legend
plt.xlabel('Time')
plt.ylabel('Probability')
plt.legend(loc='lower left',prop={'size':7.5})

# show the plot
plt.show()
```
